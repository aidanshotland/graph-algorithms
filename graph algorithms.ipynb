```python
# A class for directed graph object
class DGraph:
    def __init__(self, edges, n):
 
        # A list of lists to represent an adjacency list
        self.adjList = [[] for _ in range(n)]
 
        # add edges to the directed graph
        for (src, dest) in edges:
            # add an edge from source to destination
            self.adjList[src].append(dest)
```

```python
# Write your DFS_enhanced function to record finish time of each vertex
# You can modify arguments and define other functions if it needs
def DFS_enhanced(graph, v, discovered, departure, time,stack):
    #since it was visted call it True
    discovered[v] = True
    #takes a turn
    time[0]+=1
    #loop through adjacent values
    for i in graph.adjList[v]:
        #if not found recurse again 
        if not discovered[i]:
            DFS_enhanced(graph, i, discovered, departure, time,stack)
    #store the time it took
    departure[v] = time[0]
    #add another turn and put in stack
    time[0]+=1
    stack.append(v)
```

```python
# Write your own doTopologicalSort function to print vertex in a topological order
def doTopologicalSort(graph, n):
    #new disovered array all filled with False
    discovered = [False] *n
    #departure array, default value -1
    departure = [-1]*n
    #how many times it takes 
    time = [0]

    stack=[]
    #loops through; if disocvored[v] is False call DFS
    for v in range(n):
        if not discovered[v]:
            DFS_enhanced(graph,v,discovered,departure,time,stack)

    #creates list full of stack values along with the time for each
    t_list = [(stack[i], departure[stack[i]]) for i in range(n)]
    #reverses and prints
    t_list.reverse()
    for i in t_list:
        print(i[0], " ",  end="")
```
